```
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
```

# **MNIST(Question 1)**

```
data=load_digits()
```

```
x=data.data
y=data.target
```

```
print("x_shape=",x.shape)
print("y_shape=",y.shape)
```

```
print(x)
```

```
print(y)
```

```
km=KMeans(n_clusters=10)
pred=km.fit_predict(x)
```

```
from scipy import stats
label = np.zeros_like(predict)
for j in range(10):
    a= (pred==j)
    label[a] = stats.mode(y[a])[0]
```

```
print(a)
```

```
from sklearn.metrics import confusion_matrix,accuracy_score
```

# **Accuracy Score**

```
score=accuracy_score(y,label)
print("score=",score)
```

# **Confusion Matrix**

```
matrix=confusion_matrix(y,label)
print("\nconfusion matrix=\n",matrix)
```

# **IMAGE(Question 2)**

```
from sklearn.datasets import load_sample_image
```

# china image

```
china=load_sample_image("china.jpg")
axis = pl.axes(xticks=[], yticks=[])
axis.imshow(china)
```

```
print(china.shape)
```

```
x,y,z=china.shape
china_2d=china.reshape(x*y,z)
print(china_2d.shape)
```

```
km_cluster=KMeans(n_clusters=7)
pred=km_cluster.fit_predict(china_2d)
```

```
cluster_centers = km_cluster.cluster_centers_
cluster_labels = km_cluster.labels_
```

```
pl.figure(figsize = (10,10))
pl.imshow(cluster_centers[cluster_labels].reshape(x, y, z));
```
